## Kode Lama

```python
import os
import shutil

# Direktori sumber dan tujuan
srcDir = r"D:\Peta\Peta WB BARU 2024"
desDir = r"D:\Peta\BS Susenas Maret 2025"

# Daftar string yang harus dicari
reqString = [
    "6309070006011B",
    "6309070007006B",
    "6309080030004B",
    "6309080010002B",
    "6309080027004B",
    "6309080031001B",
    "6309080036003B",
    "6309081003001B",
    "6309081007001B",
    "6309090010001B",
    "6309100003004B",
    "6309100005002B",
    "6309100010006B",
    "6309100011008B",
    "6309100015002B",
    "6309110003001B",
    "6309110006002B",
    "6309110006006B",
    "6309110009001B"
]

# Buat folder tujuan jika belum ada
os.makedirs(desDir, exist_ok=True)

# Loop melalui file di direktori sumber
for root, dirs, files in os.walk(srcDir):
    for file in files:
        # Cek apakah salah satu string ada di nama file
        if any(reqString in file for reqString in reqString):
            source_file_path = os.path.join(root, file)
            destination_file_path = os.path.join(desDir, file)
            
            try:
                # Salin file ke direktori tujuan
                shutil.copy(source_file_path, destination_file_path)
                print(f"File '{file}' telah disalin ke '{desDir}'.")
            except Exception as e:
                print(f"Terjadi kesalahan saat menyalin file '{file}': {e}")

print("Proses penyalinan selesai.")
```

### Kode Awal

```python
import os
import shutil

# Meminta input direktori sumber dari pengguna
srcDir = input("Masukkan direktori asal: ")

# Meminta input direktori tujuan dari pengguna
desDir = input("Masukkan direktori tujuan: ")

# Meminta input string yang akan dicari dari pengguna dan mengubahnya menjadi list
reqStrings = input("Masukkan string yang hendak dicari (pisahkan dengan koma): ")
reqStrings = [s.strip() for s in reqStrings.split(',')]

# Buat folder tujuan jika belum ada
os.makedirs(desDir, exist_ok=True)

print("\n--- Proses Penyalinan Dimulai ---")

# Loop melalui file di direktori sumber
for root, dirs, files in os.walk(srcDir):
    for file in files:
        # Cek apakah salah satu string ada di nama file
        if any(reqString in file for reqString in reqStrings):
            source_file_path = os.path.join(root, file)
            destination_file_path = os.path.join(desDir, file)
            
            try:
                # Salin file ke direktori tujuan
                shutil.copy(source_file_path, destination_file_path)
                print(f"File '{file}' telah disalin ke '{desDir}'.")
            except Exception as e:
                print(f"Terjadi kesalahan saat menyalin file '{file}': {e}")

print("--- Proses penyalinan selesai. ---")
```
